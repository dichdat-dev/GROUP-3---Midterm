```python
import yaml
import os

def create_data_yaml(path_to_classes_txt, path_to_data_yaml):
    if not os.path.exists(path_to_classes_txt):
        print(f'classes.txt file not found at {path_to_classes_txt}')
        return
    with open(path_to_classes_txt, 'r') as f:
        classes = [line.strip() for line in f if line.strip()]
    number_of_classes = len(classes)


    data = {
        'path': '/kaggle/input/candy-dataset',  # đường dẫn gốc đến dataset
        'train': 'train/images',
        'val': 'validation/images',
        'nc': number_of_classes,
        'names': classes
    }

    with open(path_to_data_yaml, 'w') as f:
        yaml.dump(data, f, sort_keys=False)
    print(f'✅ Created config file at {path_to_data_yaml}\n')

# Gọi hàm
path_to_classes_txt = '/kaggle/input/candy-dataset/classes.txt'
path_to_data_yaml = '/kaggle/working/data.yaml'

create_data_yaml(path_to_classes_txt, path_to_data_yaml)

# Hiển thị nội dung
print(' File contents:\n')
!cat /kaggle/working/data.yaml
```

```python

yaml_path = '/kaggle/working/data.yaml'

with open(yaml_path, 'r') as file:
    data = file.read()

# Sửa file nội dung của data.yaml
data = data.replace('/kaggle/input/candy-dataset', '/kaggle/working/data')

# Ghi lại vào file
with open(yaml_path, 'w') as file:
    file.write(data)
```
